```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path("..")
APP_DIR = BASE_DIR / "app"
PAGE_DIR = APP_DIR / "pages"
SAVE_DIR = BASE_DIR / "Reports" / "Model_Reports"
SAVE_DIR.mkdir(parents=True,exist_ok=True)

dashboard_df = pd.DataFrame([
    ["Main Dashboard",APP_DIR / "main.py","Project overview and KPIs"],
    ["Utils File",APP_DIR / "utils.py","Reusable data loading functions"],
    ["Sales Dashboard",PAGE_DIR / "1_Sales_Dashboard.py","Sales KPIs and trends"],
    ["Customer Dashboard",PAGE_DIR / "2_Customer_Dashboard.py","Segments and churn overview"],
    ["Forecast Dashboard",PAGE_DIR / "3_Forecast_Dashboard.py","SKU demand trend and model summary"],
    ["Inventory Dashboard",PAGE_DIR / "4_Inventory_Dashboard.py","EOQ, safety stock and reorder point"],
    ["Monitoring Dashboard",PAGE_DIR / "5_Monitoring_Dashboard.py","Drift score and retraining logs"]
],columns=["Page","Path","Purpose"])

dashboard_df["Status"] = dashboard_df["Path"].apply(lambda x:"✅ Available" if x.exists() else "❌ Missing")
dashboard_df["Path"] = dashboard_df["Path"].astype(str)

dashboard_df.to_csv(SAVE_DIR / "21-Streamlit_Dashboard_Final_Check.csv",index=False)

dashboard_df
```
