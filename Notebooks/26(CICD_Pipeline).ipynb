```python
from pathlib import Path

Path("../.github/workflows").mkdir(parents=True,exist_ok=True)
```

```python
%%writefile ../.github/workflows/ci-cd.yml

name: RetailPulse CI-CD Pipeline

on:
  push:
    branches: [ main ]
  pull_request:
    branches: [ main ]

jobs:
  build-check:
    runs-on: ubuntu-latest

    steps:
      - name: Checkout repository
        uses: actions/checkout@v4

      - name: Set up Python
        uses: actions/setup-python@v5
        with:
          python-version: "3.12"

      - name: Upgrade pip
        run: python -m pip install --upgrade pip

      - name: Install dependencies
        run: pip install -r requirements.txt

      - name: Check project files
        run: |
          test -f README.md
          test -f requirements.txt
          test -f Dockerfile
          test -f app/main.py
          test -f app/utils.py
          test -f app/pages/1_Sales_Dashboard.py
          test -f app/pages/2_Customer_Dashboard.py
          test -f app/pages/3_Forecast_Dashboard.py
          test -f app/pages/4_Inventory_Dashboard.py
          test -f app/pages/5_Monitoring_Dashboard.py

      - name: Import check
        run: python -m py_compile app/main.py app/utils.py
```

```python
from pathlib import Path
import pandas as pd

BASE_DIR = Path("..")
SAVE_DIR = BASE_DIR / "Reports" / "Model_Reports"
SAVE_DIR.mkdir(parents=True,exist_ok=True)

cicd_files = {
    "GitHub Actions Folder":BASE_DIR / ".github" / "workflows",
    "CI Pipeline File":BASE_DIR / ".github" / "workflows" / "ci-cd.yml",
    "README.md":BASE_DIR / "README.md",
    "requirements.txt":BASE_DIR / "requirements.txt",
    "Dockerfile":BASE_DIR / "Dockerfile",
    "Streamlit Main App":BASE_DIR / "app" / "main.py"
}

cicd_df = pd.DataFrame([
    [name,path,"✅ Available" if path.exists() else "❌ Missing"]
    for name,path in cicd_files.items()
],columns=["Component","Path","Status"])

cicd_df["Path"] = cicd_df["Path"].astype(str)
cicd_df.to_csv(SAVE_DIR / "26-CICD_Pipeline_Check.csv",index=False)

cicd_df
```
