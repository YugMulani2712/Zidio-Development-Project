```python
from pathlib import Path

Path("../k8s").mkdir(parents=True,exist_ok=True)
print("k8s folder created.")
```

```python
%%writefile ../k8s/deployment.yaml

apiVersion: apps/v1
kind: Deployment
metadata:
  name: retailpulse-deployment
spec:
  replicas: 1
  selector:
    matchLabels:
      app: retailpulse
  template:
    metadata:
      labels:
        app: retailpulse
    spec:
      containers:
        - name: retailpulse-container
          image: retailpulse:latest
          imagePullPolicy: IfNotPresent
          ports:
            - containerPort: 8501
```

```python
%%writefile ../k8s/service.yaml

apiVersion: v1
kind: Service
metadata:
  name: retailpulse-service
spec:
  type: NodePort
  selector:
    app: retailpulse
  ports:
    - protocol: TCP
      port: 8501
      targetPort: 8501
      nodePort: 30085
```

```python
from pathlib import Path
import pandas as pd

SAVE_DIR = Path("../Reports/Model_Reports")
SAVE_DIR.mkdir(parents=True,exist_ok=True)

k8s_df = pd.DataFrame([
    ["deployment.yaml",Path("../k8s/deployment.yaml")],
    ["service.yaml",Path("../k8s/service.yaml")]
],columns=["File","Path"])

k8s_df["Status"] = k8s_df["Path"].apply(lambda x:"✅ Available" if x.exists() else "❌ Missing")
k8s_df["Path"] = k8s_df["Path"].astype(str)

k8s_df.to_csv(SAVE_DIR / "25-Kubernetes_Setup_Check.csv",index=False)

k8s_df
```
